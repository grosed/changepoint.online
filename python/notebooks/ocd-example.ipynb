```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
N = 50000
mu = 0.08
Y = np.concatenate((np.random.normal(loc=0.0, scale=1.0, size=N),np.random.normal(loc=mu, scale=1.0, size=N)))
X = list(Y)
```

```python
method = ocd_mwu(2000,2000)
```

```python
score = list()
for x in X :
    method.push(x)
    score.append(method.z)
```

```python
plt.plot([s for s in score if s != None])
```
